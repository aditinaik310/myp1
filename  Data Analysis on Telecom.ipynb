```python
#catagorical and numerical
categorical=[]
numerical=[]

for i in df.columns:
    if df[i].dtypes=='O':
        categorical.append(i)
    else:
        numerical.append(i)
        
print('categorical:',categorical)
print(' numerical:', numerical)
```

```python
categorical=['gender','SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod','Churn']
numerical =[ 'tenure', 'MonthlyCharges']
```

```python
import matplotlib.pyplot as plt #import matplotlib
for i in categorical:
    print(df[i].unique())
    print(df[i].value_counts())
    df[i].value_counts().plot(kind='bar')
    plt.show()
```

```python
for i in numerical:
    df[i].plot(kind='kde')
    plt.show()
```

```python
df.groupby('gender')['tenure'].mean() # groupby using gender and tenure to find mean
```

```python
df.groupby('gender')['tenure'].mean().plot(kind="pie") #plot the graph
```

```python
categorical=['gender','SeniorCitizen','Partner','Dependents','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod','Churn']
numerical =[ 'tenure', 'MonthlyCharges']
```

```python
for i in categorical: #to perform avarage,max,min
    for j in numerical:
        print(f'The average {j} for each category {i}  {df.groupby(i)[j].mean()}')
        print(f'the max {j} for each category {i}  {df.groupby(i)[j].max()}')
        print(f'the min  {j} for each category {i} {df.groupby(i)[j].min()}')
```
